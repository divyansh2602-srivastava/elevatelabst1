Task2: Exploratory Data Analysis(EDA)

Objective: Understand data using statistics and visualization

```python
#Import Libraries

import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
#Load the datset

df = pd.read_csv(r"c:\Users\divya\Downloads\Titanic-Dataset (1).csv")
df.head()
```

```python
#Summary Statistics 

df.describe()
df.info()
df.isnull().sum()
```

Univariate Analysis

```python
#Histogram For distribution 

df['Age'].hist(bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#Boxplot for Outliers Detection

sns.boxplot(x=df['Fare'])
plt.title("Fare Boxplot")
plt.show()
```

Multivariable Ananlysis

```python
#Pairplot

sns.pairplot(df[['Age', 'Fare', 'Survived']])
plt.show()
```
